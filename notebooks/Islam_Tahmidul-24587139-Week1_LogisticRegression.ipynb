# 2. Prepare data

```python
df_train_cleaned = df_train.copy()
df_test_cleaned = df_test.copy()
```

Dropping columns from dataset because of excessive missing values

```python
df_train_cleaned = df_train_cleaned.drop(['Rec_Rank', 'pick', 'dunks_ratio'], axis = 1 )
df_test_cleaned = df_test_cleaned.drop(['Rec_Rank', 'pick', 'dunks_ratio'], axis = 1)
```

Dropping columns because the range of categorical data is too high to encode into numbers or irrelevant

```python
df_train_cleaned = df_train_cleaned.drop(['team', 'conf', 'ht', 'yr', 'type','num', 'player_id'], axis = 1 )
df_test_cleaned = df_test_cleaned.drop(['team', 'conf', 'ht', 'yr', 'type', 'num'], axis = 1) 
```

```python
df_train_cleaned.info()
```

```python
print(df_train_cleaned.shape)
print(df_test_cleaned.shape)
```

Performing imputations on columns that have null values

```python
for i in range(0,54):
    if df_train_cleaned.iloc[:, i].isna().sum() != 0:
        mean_value = df_train_cleaned.iloc[:, i].mean()
        df_train_cleaned.iloc[:, i].fillna(mean_value, inplace=True)
        
    if df_test_cleaned.iloc[:, i].isna().sum() != 0:
        mean_value = df_test_cleaned.iloc[:, i].mean()
        df_test_cleaned.iloc[:, i].fillna(mean_value, inplace=True)    
    
        
```

Storing all the player ids from the Test data set

```python
player_id = df_test_cleaned.pop('player_id')
```

# 3. Data Splitting


```python
y_data = df_train_cleaned.pop('drafted')
```

```python
X_data = df_train_cleaned.copy()
X_test = df_test_cleaned.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_data, y_data, test_size=0.2, random_state=8)
```

# 4. Data Scaling/Standardising

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from joblib import dump
dump(scaler, '../models/scaler.joblib')
```

```python
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train))
X_val_scaled = pd.DataFrame(scaler.fit_transform(X_val))
X_test_scaled = pd.DataFrame(scaler.fit_transform(X_test))
```

```python
X_train.to_csv('../data/processed/X_train_scaled.csv', index=False)
X_val.to_csv('../data/processed/X_val_scaled.csv', index=False)
X_test.to_csv('../data/processed/X_test_scaled.csv', index=False)
y_train.to_csv('../data/processed/y_train.csv', index=False)
y_val.to_csv('../data/processed/y_val.csv', index=False)
```

# 5. Applying PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(.99)
dump(pca, '../models/pca.joblib')
```

```python
pca.fit(X_train)
```

```python
pca.n_components_
```

```python
X_train_pca = pca.transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)
```

# 6. Apply Polynomial Transformation

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(2)
```

```python
dump(poly, '../models/poly.joblib')
```

```python
X_train_pca = poly.fit_transform(X_train_pca)
```

```python
X_val_pca = poly.transform(X_val_pca)
X_test_pca = poly.transform(X_test_pca)
```

# 7. Train Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
clf = reg.fit(X_train_pca, y_train)
```

```python
from joblib import dump

dump(reg,  '../models/log_reg.joblib')
```

```python
y_train_preds = reg.predict(X_train_pca)
y_val_preds = reg.predict(X_val_pca)
y_test_preds = reg.predict(X_test_pca)
```
